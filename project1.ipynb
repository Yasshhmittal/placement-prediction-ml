```python
# Steps

# 0. Preprocess + EDA + Feature Selection
# 1. Extract input and output cols
# 2. Scale the values
# 3. Train test split
# 4. Train the model
# 5. Evaluate the model/model selection
# 6. Deploy the model
```

```python
df = df.iloc[:,1:] # isse unnamed wala column delete hojayga
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
x = df.iloc[:, 0:2]
y = df.iloc[:,-1]
```

```python
x
```

```python
y.shape
```

```python
%pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.1)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test = scaler.transform(x_test)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
# model training
clf.fit(x_train, y_train)
```

```python
# thorugh this we can predict the the output 
y_pred = clf.predict(x_test)
```

```python
y_test
```

```python
# through this class we can check the accuracy of our modal
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
# to visualize ki what pattern our model have seen in order to predict outcome
# we use decision boundary
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x_train, y_train.values, clf=clf, legend = 2)
```
